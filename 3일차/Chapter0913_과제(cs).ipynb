[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

얕은 복사 : 객체를 복사할 때 하위 객체의 참조를 공유합니다 복사된 객체가 동일한 하위 객체를 참조하므로 하위 객체의 변경이 양쪽으로 영향이 미칠 수 있습니다

깊은 복사 : 객체를 복사할 때 하위 객체도 새로 복사하여 독립적인 복사본을 생성합니다.
복사된 객체와 원본객체가 서로 독립적으로 하위 객체를 가지므로 한 객체의 하위 객체를 변경해도 다른 객체에는 영향이 미치지 않습니다
